A journal preset sets the base rules for your figure: fonts, sizes, line weights, and colors. An overlay lets you adjust one aspect of that without changing the base journal settings.

For example, you can use the `nature` journal preset and add the `minimal` overlay to strip the top and right spines, giving an editorial look while keeping all of Nature's typography rules.

**What you will learn:**
- What overlays are and how they work
- Context overlays: adjust scale for notebooks, presentations, and editorial figures
- Color overlays: swap the color palette without changing journal rules
- Plot-type overlays: optimise rcParams for bar charts and scatter plots
- Combining multiple overlays in one call
- How to list all available overlays

````python
import matplotlib.pyplot as plt
import numpy as np

import plotstyle

rng = np.random.default_rng(42)
x = np.linspace(0, 2 * np.pi, 100)
````

## 1. How Overlays Work

Pass overlay names alongside a journal key in `plotstyle.use()`. Overlays are applied in order on top of the journal's base settings.

```python
# Journal only
with plotstyle.use("nature") as style:
    ...

# Journal + one overlay
with plotstyle.use(["nature", "minimal"]) as style:
    ...

# Journal + multiple overlays
with plotstyle.use(["nature", "minimal", "okabe-ito"]) as style:
    ...
```

Overlays are grouped into five categories:

| Category | What it changes | Examples |
|---|---|---|
| `color` | Color cycle / palette | `okabe-ito`, `tol-bright`, `safe-grayscale` |
| `context` | Scale for the medium | `minimal`, `notebook`, `presentation`, `high-vis` |
| `rendering` | LaTeX and grid settings | `no-latex`, `grid`, `latex-sans`, `pgf` |
| `plot-type` | Chart-specific rcParams | `bar`, `scatter` |
| `script` | Non-Latin font support | `cjk-simplified`, `russian`, `turkish` |

## 2. Context Overlay: `minimal`

The `minimal` overlay strips the top and right axes spines for a clean, editorial look. It keeps all of the journal's fonts, sizes, and line weights.

````python
with plotstyle.use(["nature", "minimal"]) as style:
    fig, ax = style.figure(columns=1)
    for i, color in enumerate(style.palette(n=3)):
        ax.plot(x, np.sin(x + i * 0.5), color=color, label=f"Series {i + 1}")
    ax.set_xlabel("Phase (rad)")
    ax.set_ylabel("Amplitude")
    ax.legend()
    plt.show()
    plt.close(fig)
````

## 3. Context Overlay: `notebook`

The `notebook` overlay increases the figure size and font sizes so your figure is easy to read inline in a Jupyter notebook. Use `plt.subplots()` here instead of `style.figure()` so the overlay's larger figure size takes effect.

````python
with plotstyle.use(["nature", "notebook"]) as style:
    # plt.subplots() picks up the overlay's figure.figsize rcParam
    fig, ax = plt.subplots()
    ax.plot(x, np.sin(x), label="sin(x)")
    ax.plot(x, np.cos(x), label="cos(x)")
    ax.set_xlabel("Phase (rad)")
    ax.set_ylabel("Amplitude")
    ax.legend()
    plt.show()
    plt.close(fig)
````

## 4. Context Overlay: `presentation`

The `presentation` overlay uses large text and thick lines, making figures legible when projected on a screen. Combine it with any journal preset.

````python
with plotstyle.use(["nature", "presentation"]) as style:
    fig, ax = plt.subplots()
    ax.plot(x, np.sin(x), label="sin(x)")
    ax.plot(x, np.cos(x), label="cos(x)")
    ax.set_xlabel("Phase (rad)")
    ax.set_ylabel("Amplitude")
    ax.legend()
    plt.show()
    plt.close(fig)
````

## 5. Color Overlays: Swap the Palette

Color overlays replace the default color cycle with an alternative palette. This is useful when you need a specific accessible palette, or when you want to use a consistent palette across multiple journals.

| Overlay | Palette | Good for |
|---|---|---|
| `okabe-ito` | 8-color colorblind-safe | Qualitative data with up to 8 groups |
| `tol-bright` | Paul Tol's 7-color bright | High contrast on light backgrounds |
| `tol-muted` | Paul Tol's 10-color muted | Dense figures with many groups |
| `safe-grayscale` | 6-step grayscale | Black-and-white print |

The color overlay only changes the color cycle. All other journal settings stay the same.

````python
# Compare default Nature palette vs okabe-ito overlay on the same journal
fig, axes = plt.subplots(1, 2, figsize=(9, 3))

# Left: Nature's default palette
with plotstyle.use("nature") as style:
    colors = style.palette(n=4)
    for i, c in enumerate(colors):
        axes[0].plot(x, np.sin(x + i * 0.5), color=c, label=f"S{i + 1}")
    axes[0].set_title("Nature default palette")
    axes[0].set_xlabel("x")
    axes[0].legend()

# Right: Same journal with okabe-ito color overlay
with plotstyle.use(["nature", "okabe-ito"]) as style:
    colors = style.palette(n=4)
    for i, c in enumerate(colors):
        axes[1].plot(x, np.sin(x + i * 0.5), color=c, label=f"S{i + 1}")
    axes[1].set_title("Nature + okabe-ito overlay")
    axes[1].set_xlabel("x")
    axes[1].legend()

fig.tight_layout()
plt.show()
plt.close(fig)
````

## 6. Plot-Type Overlays: `bar` and `scatter`

These overlays tune rcParams for specific chart types:

- `bar`: Removes markers and sets a narrow patch border, which is recommended for bar charts.
- `scatter`: Removes line connectors and adjusts marker size defaults for scatter plots.

````python
categories = ["Control", "Treatment A", "Treatment B", "Treatment C"]
values = [1.00, 1.34, 0.87, 1.56]
errors = [0.05, 0.08, 0.06, 0.09]

with plotstyle.use(["nature", "bar"]) as style:
    fig, ax = style.figure(columns=1)
    colors = style.palette(n=len(categories))
    ax.bar(categories, values, yerr=errors, color=colors, capsize=3)
    ax.set_ylabel("Relative expression")
    ax.axhline(1.0, color="black", linewidth=0.5, linestyle="--")
    plt.show()
    plt.close(fig)
````

## 7. Combining Multiple Overlays

You can stack overlays from different categories. For example: a journal preset, a color overlay, and a context overlay together.

Overlays are applied in order, so the last overlay wins if two overlays set the same rcParam.

````python
# ieee journal + okabe-ito color palette + minimal spines
with plotstyle.use(["ieee", "okabe-ito", "minimal"]) as style:
    fig, ax = style.figure(columns=1)
    for i in range(4):
        ax.plot(x, np.sin(x + i * 0.5), label=f"Series {i + 1}")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.legend()
    plt.show()
    plt.close(fig)
````

## 8. List All Available Overlays

`plotstyle.list_overlays()` returns every overlay key. Pass `category=` to filter by type.

````python
# All overlays, grouped by category
all_overlays = plotstyle.list_overlays()
print(f"Total overlays available: {len(all_overlays)}\n")

for category in ["context", "color", "plot-type", "rendering", "script"]:
    keys = plotstyle.list_overlays(category=category)
    print(f"[{category}]")
    for key in keys:
        ov = plotstyle.overlay_registry.get(key)
        print(f"  {key:<22} {ov.description[:55]}")
    print()
````
